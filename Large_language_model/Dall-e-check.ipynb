```python
# 2- Display using PIL - Python Imaging Library 
from io import BytesIO
import requests
from PIL import Image

response = openai.Image.create(
    prompt = "create an image of an old computer",
    size = "256x256",
    n = 1,
    response_format = "url"
)
response = requests.get(response["data"][0]["url"])
##use bytesio to convert the  response content  to the function
Image.open(BytesIO(response.content))
```

```python
# 3- Response format "b64_json"+
##b64.json encoding  will represent  something binary  like image  as a text
import base64
from IPython.display import display
from PIL import Image

response = openai.Image.create(
    prompt = "create an image of an old computer",
    size = "512x512",
    n = 1,
    response_format = "b64_json"
)

imagem = response["data"][0]["b64_json"]
imagem = base64.b64decode(imagem)

with open("computer.png", "wb") as f:
    f.write(imagem)

display(Image.open("computer.png"))
```

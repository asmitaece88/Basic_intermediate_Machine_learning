```python
import keras 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

# y = mx + b +noise

```python
m=2
b=3
x=np.linspace(0 , 50,100)
```

```python
np.random.seed(101)
noise=np.random.normal(loc=0,scale=4,size=len(x))
```

```python
noise
```

```python
x
```

```python
y=2*x+b+noise
```

```python
y
```

```python
plt.plot(x,y,'*')
plt.show()
```

# now make use of keras  package to find the best fit line ## linear regression

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense##dense layer means  all the neurons from previus layer are conected to the next layer 
#The dense layer is a neural network layer that is connected deeply, which means each neuron in the dense layer receives input from all neurons of its previous layer.
```

```python
model =Sequential()
## input_dim =1  , becauase we have 1  feature
#units: Positive integer, dimensionality of the output space.
model.add(Dense(units=4,input_dim =1 ,activation = 'relu' )) 
## next , we keep on adding mroe and mroe dense layers 
model.add(Dense(units=4,activation = 'relu' ))
## finally the output should be 1 , hence the output should be 1 
model.add(Dense(units=1 , activation ='linear'))
```

# Once the model is created, you can config the model with losses and metrics with model.compile(), 
# train the model with model.fit(), or use the model to do prediction with model.predict().

```python
model.compile(loss='mse',optimizer='adam')
```

```python
model.fit(x,y,epochs=200)
```

# as epoch increases , we see the loss paraeter gets decreased or optmized 

```python
model.history.history
## this is a dictionary 
```

```python
model.summary()
```

```python
model.history.history['loss']
```

```python
loss=model.history.history['loss']
epochs = range(len(loss))
plt.plot(epochs , loss)
```

```python
# hence the loss is decreasng as epochs increases 
```

```python
x_for_predictions =np.linspace(0,50,100)

y_pred= model.predict(x_for_predictions)
```

```python
plt.plot(x,y,".")
plt.plot(x_for_predictions,y_pred,'r')
```

```python
#  next to find out error

from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error (y,y_pred)
```

# character counts

```python
def char_counts(strin):
    s=strin.split()
    print(s)
    strin=''.join(s)
    return len(strin)
```

```python
char_counts('asmita is a good girl')## here the empty spaces are left out 
```

```python
df['char_counts']=df['twitts'].apply(lambda x : char_counts(str(x)))
```

```python
df.sample(5)
```

```python
# 1 thing to observe is  character counts ncreases as when the word count increases 
```

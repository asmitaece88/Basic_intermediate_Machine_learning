# how  to downlaod  any pre training dataloader dataset and apply torchvision 

```python
#!pip install torch torchvision torchaudio
#reference datset , will be  downloaded from Fashion Mnist 

import torch
from torch.utils.data import Dataset
from torchvision import datasets
from torchvision.transforms import ToTensor
from torch.utils.data.sampler import SubsetRandomSampler
import matplotlib.pyplot as plt
import numpy as np
```

```python
##create the training datset 

training_data = datasets.FashionMNIST(
    root="data",
    train=True,##
    download=True,
    transform=ToTensor()
)
```

```python
##create a testing datset 

test_data = datasets.FashionMNIST(
    root="data",
    train=False,##
    download=True,
    transform=ToTensor()
)
```

```python
## create a validation sample from the training dataset 
list(range(len(training_data)))
```

```python
indices=list(range(len(training_data)))
```

```python
np.random.shuffle(indices)
split=int(np.floor(0.2*len(training_data)))
```

```python
split
```

```python
train_sample=SubsetRandomSampler(indices[:split])
validation_sample =SubsetRandomSampler(indices[split:])
```

```python
## data loader 
trainloader =torch.utils.data.DataLoader(training_data,sampler=train_sample,batch_size=64)
validloader = torch.utils.data.DataLoader(training_data,sampler=validation_sample,batch_size=64)
testloader = torch.utils.data.DataLoader(test_data,batch_size=64 , shuffle=True)
```

```python
# create a data iterator
%matplotlib inline
dataiter = iter(trainloader)
print(dataiter)
images , labels= next(dataiter)
```

```python
images, labels
```

```python
#plot the data 

fig= plt.figure(figsize=(15,5))
for i in np.arange(20):
    ax = fig.add_subplot(4, int(20/4), i+1, xticks=[], yticks=[])
    ax.imshow(np.squeeze(images[i]), cmap='gray')
    ax.set_title(labels[i].item())
    fig.tight_layout()
```

# Initialize the rail

```python
colang_content = """
define user express greeting
  "hello"
  "hi"

define bot express greeting
  "Hello there!! Can I help you today?"

define flow hello
  user express greeting
  bot express greeting
"""
yaml_content = """
models:
- type: main
  engine: openai
  model: gpt-4o
"""
```

```python
from nemoguardrails import LLMRails, RailsConfig

# initialize rails config
config = RailsConfig.from_content(
    colang_content=colang_content,
    yaml_content=yaml_content
)
```

```python
config
```

```python
rails = LLMRails(config=config)# requires variable HF_token for access
```

```python
res = await rails.generate_async(prompt="Hello")
```

```python
print(res)
```

# lets take the dynamic variable

```python
colang_content_new = """
define user express greeting
    "hello"
    "hi"

define bot express greeting
    "Hello there!! Can I help you today?"

define bot personal greeting
    "Hello $username, nice to see you again!"

define flow hello
    user express greeting
    if $username
        bot personal greeting
    else
        bot express greeting
"""

```

```python
config = RailsConfig.from_content(
    yaml_content=yaml_content, colang_content=colang_content_new
)
```

```python
rails = LLMRails(config=config)
```

```python
messages
```

```python
res_2 = await rails.generate_async(messages=messages)
res_2
```

```python
res_2
```

```python
messages += [
    res_2,
    {"role": "user", "content": "I'm Asmita"}
]
```

```python
messages
```

```python
res_3 = await rails.generate_async(messages=messages)
res_3
```
